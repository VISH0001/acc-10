```python
import mpmath as mp
import matplotlib.pyplot as plt
from concurrent.futures import ProcessPoolExecutor, as_completed
import multiprocessing

mp.mp.dps = 200

# =========================================================
# FIXED lp
# =========================================================
lp = mp.mpf('1e-8')

# range of R1 values
R1_values = [mp.mpf(i) for i in range(1,21)]

results = []

# =========================================================
# BISECTION METHOD
# =========================================================
def bisection(f,a,b,tol=1e-20,max_iter=100):

    fa=f(a)
    fb=f(b)

    if fa*fb>0:
        return None

    for _ in range(max_iter):

        c=(a+b)/2
        fc=f(c)

        if abs(fc)<tol:
            return c

        if fa*fc<0:
            b=c
            fb=fc
        else:
            a=c
            fa=fc

    return (a+b)/2


# =========================================================
# LOOP OVER R1
# =========================================================
for R1 in R1_values:

    print("\n====================================")
    print("Running for R1 =",R1)
    print("====================================\n")

    R2 = mp.mpf('0')

    omega_max = mp.mpf('10')
    omega_step = mp.mpf('0.01')

    n_max = 6

    # J cutoff
    Jmax = int(mp.floor(2*R1/lp))


    # =========================================================
    # SAMPLE J VALUES
    # =========================================================
    def generate_sampled_J_values(start,stop,step_factor):

        J_values=[]
        current=start
        step=1

        while current<stop:

            next_limit=min(current*step_factor,stop)

            J_values+=list(range(int(current),int(next_limit),int(step)))

            current*=step_factor
            step*=step_factor

        return sorted(set(J_values))


    sampled_Js = generate_sampled_J_values(1,Jmax+1,10)


    # =========================================================
    # WAVE EQUATION
    # =========================================================
    def equation(omega,J):

        xi=(2*R1**2)/(R1-R2)*mp.log((R1-R2)/(2*R1**2)*lp)

        order=(2j*R1**2*omega)/(R1-R2)

        arg=(2*mp.e**(((R1-R2)*xi)/(2*R1**2))
            *mp.sqrt(J*(J+1))*R1)/(R1-R2)

        val=mp.sinh((2*mp.pi*R1**2*omega)/(R1-R2))*mp.besselk(order,arg)

        return mp.re(val)


    # =========================================================
    # ROOT REFINEMENT
    # =========================================================
    def refine_root(w1,w2,J):

        f=lambda w: equation(w,J)

        try:
            return mp.findroot(f,(w1,w2))
        except:
            return bisection(f,w1,w2)


    # =========================================================
    # ROOT SEARCH FOR FIXED J
    # =========================================================
    def roots_for_J(J):

        mp.mp.dps = 200

        roots=[]

        w=mp.mpf('0')
        prev_val=equation(w,J)

        while w<omega_max and len(roots)<n_max:

            w_next=w+omega_step
            val=equation(w_next,J)

            if not(mp.isnan(prev_val) or mp.isnan(val)) and prev_val*val<0:

                root=refine_root(w,w_next,J)

                if root is not None:
                    roots.append(root)

            w=w_next
            prev_val=val

        return J,roots


    # =========================================================
    # PARALLEL ROOT COMPUTATION
    # =========================================================
    all_roots={}

    workers=multiprocessing.cpu_count()

    with ProcessPoolExecutor(max_workers=workers) as executor:

        futures=[executor.submit(roots_for_J,J) for J in sampled_Js]

        for f in as_completed(futures):

            J,roots=f.result()
            all_roots[J]=roots


    # =========================================================
    # SPECTRUM PLOT
    # =========================================================
    plt.figure(figsize=(7,5))

    for n in range(n_max):

        Js=[]
        omegas=[]

        for J in sorted(all_roots):

            if len(all_roots[J])>n:

                Js.append(J)
                omegas.append(float(all_roots[J][n]))

        plt.plot(Js,omegas,marker='o')

    plt.xlabel("J")
    plt.ylabel("omega_n(J)")
    plt.title(f"Spectrum R1={R1}")
    plt.grid(True)
    plt.show()


    # =========================================================
    # THERMODYNAMICS
    # =========================================================
    beta = 4*mp.pi*R1**2/(R1-R2)

    sorted_J = sorted(all_roots)

    S_up = mp.mpf('0')
    E_up = mp.mpf('0')

    S_low = mp.mpf('0')
    E_low = mp.mpf('0')

    for i,J in enumerate(sorted_J):

        if i==len(sorted_J)-1:
            J_next=Jmax+1
        else:
            J_next=sorted_J[i+1]

        degeneracy = J_next**2 - J**2

        # ---------- upper ----------
        for omega in all_roots[J]:

            if abs(omega) < mp.mpf('1e-12'):
                continue

            expbw = mp.e**(beta*omega)
            bose = 1/(expbw-1)

            E_up += degeneracy*omega*bose
            S_up += degeneracy*(beta*omega*bose - mp.log(1-1/expbw))


        # ---------- lower ----------
        if i!=len(sorted_J)-1:

            Jp=sorted_J[i+1]

            for omega in all_roots[Jp]:

                if abs(omega) < mp.mpf('1e-12'):
                    continue

                expbw = mp.e**(beta*omega)
                bose = 1/(expbw-1)

                E_low += degeneracy*omega*bose
                S_low += degeneracy*(beta*omega*bose - mp.log(1-1/expbw))


    # =========================================================
    # GEOMETRIC NORMALIZATION
    # =========================================================
    E_geom = R1/(2*lp**2)
    S_geom = mp.pi*R1**2/(2*lp**2)

    Sup = S_up/S_geom
    Slow = S_low/S_geom

    Eup = E_up/E_geom
    Elow = E_low/E_geom


    print("\n---- Upper bound ----")
    print("S/S_geom =",Sup)
    print("E/E_geom =",Eup)

    print("\n---- Lower bound ----")
    print("S/S_geom =",Slow)
    print("E/E_geom =",Elow)

    results.append((R1,Sup,Slow,Eup,Elow))


# =========================================================
# FINAL TABLE
# =========================================================
print("\n\n========== FINAL DATA ==========")

for R1,su,sl,eu,el in results:

    print(f"R1={R1}  Sup={su}  Slow={sl}  Eup={eu}  Elow={el}")
```
